```python

img = rgb2gray(io.imread("./owl.jpeg")[...,:3])
io.imshow(img)
```

```python
'''
(1) Your own median filter algorithm
'''
def my_median_filter(image):
    newimg = image.copy()
    filtersize = 3
    margin = filtersize //2
    imageheight, imagewidth = image.shape
   
    for x in range(margin, imagewidth):
        for y in range(margin, imageheight):
            filter = np.zeros((filtersize, filtersize))
            filter_area = image[y - margin:y + margin + 1, x - margin:x + margin + 1]
            median = np.median(filter_area)
            newimg[y][x]= median
    return newimg
show_images([img , my_median_filter(img)])
```

```python
'''
(2) skimage median filter
'''
skimg = skifilter.median(img)
show_images([img , skimg])
```

```python
'''
(3) Gaussion Filters
'''
smoothed_image_0 = gaussian_filter(img, sigma=5)
show_images([img , smoothed_image_0])
```
